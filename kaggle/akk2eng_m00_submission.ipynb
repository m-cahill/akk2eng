# akk2eng M00 — Kaggle submission stub

**Self-contained** (no `src/` imports). Produces `submission.csv` under `/kaggle/working/`.

Attach the competition dataset. If `INPUT_DIR` is wrong, list `/kaggle/input` in a cell and set `INPUT_DIR` to the folder that contains `test.csv`.

```python
import os

# Competition data root (adjust if your run shows a different subfolder under /kaggle/input)
INPUT_DIR = "/kaggle/input/deep-past-initiative-machine-translation"
TEST_CSV = os.path.join(INPUT_DIR, "test.csv")
SUBMISSION_PATH = "/kaggle/working/submission.csv"

if not os.path.isfile(TEST_CSV):
    raise FileNotFoundError(
        f"Missing {TEST_CSV}. List /kaggle/input and set INPUT_DIR to the dataset folder."
    )
print("TEST_CSV:", TEST_CSV)
print("OUT:", SUBMISSION_PATH)
```

```python
import pandas as pd

DUMMY_TRANSLATION = "dummy translation"

test_df = pd.read_csv(TEST_CSV)
if "id" not in test_df.columns:
    raise ValueError("test.csv must contain an 'id' column")

translations = [DUMMY_TRANSLATION] * len(test_df)
submission = pd.DataFrame({"id": test_df["id"], "translation": translations})
submission.to_csv(SUBMISSION_PATH, index=False)

check = pd.read_csv(SUBMISSION_PATH)
assert list(check.columns) == ["id", "translation"]
print(check.head())
print("rows:", len(check))
```
